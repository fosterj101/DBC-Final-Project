## Gross Domestic Product


```python
gdp_df = pd.read_csv('/content/gdp1.tsv', sep = '\t')

#renaming, splitting, droping, and reordering columns
gdp_df[['freq','unit','metric','country']] = gdp_df['freq,unit,na_item,geo\TIME_PERIOD'].str.split(',',expand=True)
gdp_df = gdp_df.drop(columns=['freq,unit,na_item,geo\TIME_PERIOD','freq','unit'])

#removing unwanted values and labeling NaN cells
gdp_df = gdp_df.replace({'b':'','p':'','e':'',':':'',' ':''}, regex=True)
gdp_df = gdp_df.replace('',np.nan)
gdp_df.columns = gdp_df.columns.str.replace(' ', '') 

#convert df to numeric
gdp_df = gdp_df.apply(pd.to_numeric, errors='ignore')

# replacing metric codes with labels
for index in gdp_df.index:
    if gdp_df.loc[index,'metric']=='B1GQ':
        gdp_df.loc[index,'metric'] = 'gdp'
    if gdp_df.loc[index,'metric']=='P3_S13':
        gdp_df.loc[index,'metric'] = 'govt_expd'
    if gdp_df.loc[index,'metric']=='P31_S14':
        gdp_df.loc[index,'metric'] = 'hh_expd'
    if gdp_df.loc[index,'metric']=='P6':
        gdp_df.loc[index,'metric'] = 'imports'
    if gdp_df.loc[index,'metric']=='P7':
        gdp_df.loc[index,'metric'] = 'exports'
    if gdp_df.loc[index,'metric']=='D21X31':
        gdp_df.loc[index,'metric'] = 'taxes'

#creating MultiLevel index
gdp_multi = gdp_df.set_index(['metric','country'])

#show multi df
gdp_multi.head()
```

### A1 - GDP

```python
#creating gdp df from multilevel df
gdp = gdp_multi.loc['gdp'].T

recent_gdp = gdp[-1:].sort_values(by='2019', ascending=False, axis=1)

fig, ax = plt.subplots(figsize=(12,10))
ax.set_title('GDP: 2019', fontsize=20)
recent_gdp.plot.barh(ax=ax, width=10)
ax.set_xlabel('millions', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(recent_gdp, loc='upper center', bbox_to_anchor=(.5, .98),
          ncol=6, fancybox=True, shadow=True, fontsize=10)

```

## Consumer Price Index (CPI)

```python
cpi_df =  pd.read_csv('/content/cpi1.tsv', delimiter = '\t')


#renaming, splitting, and droping columns
cpi_df[['freq','unit','s_adj','metric','country']] = cpi_df['freq,unit,s_adj,indic,geo\TIME_PERIOD'].str.split(',',expand=True)
cpi_df = cpi_df.drop(columns=['freq,unit,s_adj,indic,geo\TIME_PERIOD','freq','unit','s_adj'])

#removing unwanted values and labeling NaN cells
cpi_df = cpi_df.replace({':':'',' ':'','e':'','r':'','d':''}, regex=True)
cpi_df = cpi_df.replace('',np.nan)

# replacing metric codes with labels
for index in cpi_df.index:
    if cpi_df.loc[index,'metric']=='CP-HI00':
        cpi_df.loc[index,'metric'] = 'cpi'
    if cpi_df.loc[index,'metric']=='CP-HI06':
        cpi_df.loc[index,'metric'] = 'health'
    if cpi_df.loc[index,'metric']=='CP-HI10':
        cpi_df.loc[index,'metric'] = 'education'
    if cpi_df.loc[index,'metric']=='CP-HIE':
        cpi_df.loc[index,'metric'] = 'energy'
    if cpi_df.loc[index,'metric']=='CP-HI02':
        cpi_df.loc[index,'metric'] = 'alc_tabc'
        
#convert df to numeric
cpi_df = cpi_df.apply(pd.to_numeric, errors='ignore')
        
#creating MultiLevel index
cpi_multi = cpi_df.set_index(['metric','country'])

#show df
cpi_multi.head()
```
